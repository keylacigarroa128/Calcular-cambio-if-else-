## Las Variables
1. CP ---> Costo de producto
2. CD ---> Cantidad de dinero
3. Cambio

---

## Formulas
$Cambio = CD - CP$

```python
# Para cuando es verdadero (SI)
# Calculando el cambio usando if else
# datos de entrada CD , CP
# mostrar Cambio si es falso
# si es verdadero mostrar si No alcanza
CP = int(input("Ingrese la cantidad del producto: "))
CD = int(input("Ingrese la cantidad de dinero: "))
if CP >= CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Keyla Paola Cigarroa Palacios NL = 12")
```

```python
# Para cuando es falso (NO)
# Calculando el cambio usando if else
# datos de entrada CD , CP
# mostrar Cambio si es falso
# si es verdadero mostrar si No alcanza
CP = int(input("Ingrese la cantidad del producto: "))
CD = int(input("Ingrese la cantidad de dinero: "))
if CP >= CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Keyla Paola Cigarroa Palacios NL = 12")
```
